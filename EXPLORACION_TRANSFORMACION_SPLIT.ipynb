# Procesamiento de datos vacios

## EliminaciÃ³n de datos

```python
# Eliminar filas con registros vacios
df_drop=df1.dropna(axis=0)
df_drop
```

```python
# Eliminar filas con registros vacios
df_drop=df1.dropna(axis=1)
df_drop
```

## ImputaciÃ³n de valores vacios

```python
from sklearn.impute import SimpleImputer

# Separar columnas numÃ©ricas y categÃ³ricas
columnas_numericas = df1.select_dtypes(include=['number']).columns #incluye tanto int como float
columnas_categoricas = df1.select_dtypes(exclude=['number']).columns #incluye solo object

# Imputar valores faltantes en las columnas numÃ©ricas con la media
imputacion_numerica = SimpleImputer(missing_values=np.nan, strategy='mean') #mean, median, most_frequent, "strategy='constant', fill_value=0"
df1[columnas_numericas] = imputacion_numerica.fit_transform(df1[columnas_numericas])

# Imputar valores faltantes en las columnas categÃ³ricas con el valor mÃ¡s frecuente 
imputacion_categorica = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df1[columnas_categoricas] = imputacion_categorica.fit_transform(df1[columnas_categoricas])

df1
```

# Tratamiento de datos categÃ³ricos

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df1['Estado']=lb.fit_transform(df1['Estado'].values)
df1.head(5)
```

```python
columnas_dummy=df1.columns[df1.dtypes==object]
columnas_dummy
```

```python
df1 = pd.get_dummies(df1, columns=columnas_dummy, prefix=columnas_dummy, drop_first=True) #- Esto evita la multicolinealidad
#df1 = pd.get_dummies(df1, columns=columnas_dummy, prefix=columnas_dummy)
df1 = df1.replace({True:1,False:0})
df1
```

```python
X = df1.loc[:, ~df1.columns.str.contains('Estado')]
Y = df1.loc[:, df1.columns.str.contains('Estado')]                                                     
print("SeparaciÃ³n de datos usando Pandas") 
print(X.shape, Y.shape)
```

```python
from imblearn.over_sampling import RandomOverSampler
# Crear una instancia de RandomOverSampler
ros = RandomOverSampler(sampling_strategy='auto', random_state=42)
# Aplicar RandomOverSampler para generar muestras sintÃ©ticas de las clases minoritarias
X, Y = ros.fit_resample(X, Y)
frecuencias = Y['Estado'].value_counts()
print(frecuencias)
```

```python
X
```

# Split de datos

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,Y, test_size=0.3, random_state=1,stratify=Y)
```
